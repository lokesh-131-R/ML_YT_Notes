```python
#na_values
df7=pd.read_csv('G:\\Data science\\Fortune_10.csv',na_values={"ID":4,'Inception':2010,'Growth':'7%'})
df7 # here na_values is use to make the values to nan by passing the dictnory to it 
```

```python
#isnull()
df7.isnull().sum() # to find out the null values in the dataset
```

```python
#notnull()
df7.notnull().sum() # to find out the num of total non nan values
```

```python
#dropna()
df7.dropna() # this will drop all the rows where nan values are present  
```

```python
df7.dropna(how='any') # 'how' this will also drop the rows where all the nan values are there
```

```python
# fillna
df8=df7.fillna("none") # here we are making the nan value sa none 
df8# in place of none we can give any values like interger or stiring
```

```python
df9=df7.fillna({"Expenses":'not mentioned',"Growth":'not given'})# filling the none data using the dict
df9
```

```python
df10=df7.fillna(method='ffill') # here 'ffill' means they fill the data of previous row similarly 'bfill' to fill using the next row
df10
```

```python
# replace 
df11=df7.replace(to_replace="Lamtone",value="india")# here the 'Lamtone' is replace by the 'india'.
df11# here we need not mention 'to_replace' and value ever time jst directly give the values seperated by ','
```

```python
df12=df7.replace(['india','Health'],['Lamtone','hospital']) # replaceing using the list
df12# even ffill and bfill can be used for replace
```

```python
#interpolate
df13=df7.interpolate() # this uses the linerar interpolation for filling the nan values default method used is linear
df13 # even we can give the method="linear"or 'time' or 'index' or'nearest' as per required
#even function 'limit =3' for limiting the interpolation for particular number of rows
```

```python
# loc and iloc
print(df7.loc[0,'ID']) # here we can call a particular values using the index and column name 
print(df7.iloc[0:4,0:4]) # here iloc is used to call the index value here syntax is 'df.iloc[row,column]'
```

```python
# groupby 
df14=df7.groupby("Industry") # creating the group by giving the column name 
df14.get_group("IT Services")# calling the group by particular name 
```

```python
# merging 
df15=pd.merge(df7,df8,on='Name',how='outer') # here merge based on the column 
df15 # here we are using the outer join we can use the 'right','left','inner'joins also 
```

```python
# concat()
df16=pd.concat([df7,df8],axis=1,ignore_index="True")
df16 # here using the concat to join the 2 df in row wise by giving the axis =1 and ignoreing the index
```

```python
# join
df17=df7.join(df8)
df17 # this will only work when index is same 
```

```python
# append
df18=df7.append(df8)
df18 # similar to concat
```

```python
# pivol_table
df7.pivot_table(index='Industry') # this basicaly join the similar item in the columns makes the table accondingly
```
